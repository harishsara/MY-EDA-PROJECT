1. PRICE ANALYSIS:
A) WHAT IS THE AVERAGE NIGHTLY PRICE FOR LISTING IN DIFFERENT NEIGHBOURHOODS?
B) HOW DOES THE PRICE VARY BASED ON THE PROPERTY TYPE?

```python
neighbourhood_average_price = df.groupby('neighbourhood')['price'].mean().sort_values()
print("AVERAGE NIGHTLY PRICE BY NEIGHBOURHOOD:")
print(neighbourhood_average_price)
```

```python
property_type_avg_price = df.groupby('room_type')['price'].mean().sort_values()
print("AVERAGE PROPERTY PRICE:")
print(property_type_avg_price)
```

2. Availability analysis?
A) WHAT IS THE AVERAGE OCCUPANY RATE FOR THE DIFFERENT ROOM TYPES?
B) ARE THERE SPECIFIC TIME OF THE YEAR WHEN LISTINGS ARE MORE OR LESS AVAILABILITY?

```python
df['last_review'] = pd.to_datetime(df['last_review'])
room_type_occupancy = df.groupby('room_type')['last_review'].count()/len(df)*100
print("OCCUPANCY RATE BY ROOM TYPE:")
print(room_type_occupancy)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df['month'] = df['last_review'].dt.month
df['year'] = df['last_review'].dt.year
availability_over_time = df.groupby(['year','month'])['id'].count()
availability_over_time.plot(kind='line',marker='o')
plt.title('availability over time')
plt.xlabel('year-month')
plt.ylabel('no of listing')
plt.show()
```

3.NEIGHBOURHOOD TRENDS: 
A) WHICH NEIGHBOURHOODS HAVE THE HIGHEST NO OF LISTNGS?
B) ARE THERE ANY SPECIFIC NEIGHBOURHOOD THAT RECEIVE MORE POSITIVE REVIEWS?

```python
neighbourhood_listing_count = df['neighbourhood'].value_counts().sort_values(ascending=False)
print('highest no of listing:')
print(neighbourhood_listing_count)
```

```python
positive_reviews_neighbourhood = df[df['number_of_reviews']>= 100]['neighbourhood']
positive_reviews_neighbourhood_count = positive_reviews_neighbourhood.value_counts().sort_values(ascending=False)
print('neighbourhoods with more positve reviews:')
print(positive_reviews_neighbourhood_count)
```

4. HOST CHARACTERISTICS: 
A) WHAT IS THE DISTRIBUTION OF THE NUMBER OF LISTING PER HOST?
B) DO SUPERHOSTS TREND TO HAVE HIGHER RATE LISTINGS?

```python
listings_per_host_distribution = df.groupby('host_id')['id'].count().value_counts().sort_index()
print("distribution of the number of listing per host:")
print(listings_per_host_distribution)
```

```python
listings_per_host_distribution.plot(kind='bar',width=0.5)
plt.title('ditribution of listingsper host')
plt.xlabel('no of listings per host')
plt.ylabel('number of hosts')
plt.show()
```

```python
superhosts_ratings = df.groupby('host_id')['number_of_reviews'].mean()
print("average rating for superhosts nad non superhosts:")
print(superhosts_ratings)
```

```python
df['price'] = pd.to_datetime(df['price'])
df['year'] = df['price'].dt.year
df['month'] = df['price'].dt.month
peak_seasons = [11,12]
peak_season_data = df[df['month'].isin(peak_seasons)]
demand_per_month = peak_season_data.groupby(['year','month'])['id'].count()
demand_per_month.plot(kind = 'line', marker = 'o')
plt.title('demand during peak tourist seasons')
plt.xlabel('year-month')
plt.ylabel('number of listings')
plt.show()
```

6. length of stay:
a) what is the average length of stay for guests in different neighbourhoods?
b) do certain property types attract longer or shoter stays?

```python
df['minimum_nights'] = pd.to_numeric(df['minimum_nights'], errors='coerce')
avg_length_of_stay_neighborhood = df.groupby('neighbourhood')[['minimum_nights']].mean()
avg_length_of_stay_neighborhood.plot(kind='bar', stacked=True)
plt.title('Average Length of Stay in Different Neighborhoods')
plt.xlabel('Neighborhood')
plt.ylabel('Average Length of Stay (Nights)')
plt.legend(["Minimum Nights"])
plt.show()
avg_length_of_stay_property_type = df.groupby('room_type')[['minimum_nights']].mean()
avg_length_of_stay_property_type.plot(kind='bar', stacked=True)
plt.title('Average Length of Stay for Different Property Types')
plt.xlabel('Property Type')
plt.ylabel('Average Length of Stay (Nights)')
plt.legend(["Minimum Nights"])
plt.show()
```

```python
import seaborn as sns
fig = sns.catplot(x='neighbourhood_group',y='availability_365',kind='box',hue='room_type',data=df,palette='pastel')
fig.fig.suptitle('Where is the most popular area?',fontsize=15,y=1.05)
fig.savefig('popular_area.png', bbox_inches = 'tight')
```

```python
fig2 = sns.catplot(x='neighbourhood_group',y='price',data=df,kind='bar',hue='room_type',palette='pastel')
fig2.fig.suptitle('Where is the most expensive area?',fontsize=15,y=1.05)
fig2.savefig('price_area.png', bbox_inches = 'tight')
```

```python
df2 = df[df['price']>=df['price'].quantile(0.75)]
fig3 = sns.catplot(x='neighbourhood_group',data=df2,hue='room_type',kind='count',palette='Pastel1')
fig3.fig.suptitle('Who are the most successful hosts?',fontsize=15,y=1.05)
fig3.savefig('host.png', bbox_inches = 'tight')
```
